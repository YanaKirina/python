# Чистим floors_total

```python
correlation11 = data['floor'].corr(data['floors_total'])
correlation11
```

```python
def categorize_floors_total(floors: float | None) -> str:
    if floors == 1:
        return "Одноэатжка"
    elif floors == 2:
        return "Двухэтажка"
    elif 3 <= floors < 5:
        return "3-4 этажка"
    elif floors == 5:
        return "Пятиэтажка"
    elif 5 < floors < 9:
        return "6-8 этажка"
    elif floors == 9:
        return "Девятиэтажка"
    elif floors > 9:
        return "Гигантоэтажка"

data['categorized_floors_total'] = data['floors_total'].apply(categorize_floors_total)
data
```

```python
px.box(data, x = 'ceiling_height', color = 'categorized_floors_total')
```

```python
data.groupby('categorized_floors_total')['ceiling_height'].agg(['median','mean'])
```

```python
data['ceiling_height'] = data.groupby('categorized_floors_total')['ceiling_height'].transform(lambda x: x.fillna(x.median()))

data['ceiling_height'] = (data['ceiling_height'].fillna(data['ceiling_height'].median()))
```

# Чистим Locality Name

```python
sorted_locality = sorted(list(data['locality_name'].dropna().unique()))
print('\n'.join(sorted_locality))
```

```python
def filter_locality_names(name: str) -> str:
    if not name:
        return None
    elif name.lower().startswith('деревня'): name = name.replace('деревня', 'посёлок',1)
    elif name.lower().startswith('поселок'): name = name.replace('поселок', 'посёлок', 1)
    elif name.lower().startswith('село'): name = name.replace('село', 'посёлок', 1)
    elif name.lower().startswith('посёлок городского типа'): name = name.replace('посёлок городского типа', 'посёлок', 1)
    elif name.lower().startswith('посёлок при железнодорожной станции'): name = name.replace('посёлок при железнодорожной станции', 'посёлок', 1)
    elif name.lower().startswith('посёлок станции'): name = name.replace('посёлок станции', 'посёлок', 1)
    elif name.lower().startswith('коттеджный посёлок'): name = name.replace('коттеджный посёлок', 'посёлок', 1)
    elif name.lower().startswith('городской посёлок'): name = name.replace('городской посёлок', 'посёлок', 1)
    elif name.lower().startswith('садоводческое некоммерческое товарищество'): name = name.replace('садоводческое некоммерческое товарищество', 'посёлок', 1)
    elif name.lower().startswith('садовое товарищество'): name = name.replace('садовое товарищество', 'посёлок', 1)
    return name

data['locality_name'] = data['locality_name'].apply(filter_locality_names)
```

```python
def categorize_total_area(area: float | None) -> str:
    if area < 10:
        return "Мало"
    elif 10 <= area < 15:
        return "10-15"
    elif 15 <= area < 20:
        return "15-20"
    elif 20 <= area < 25:
        return "20-25"
    elif 25 <= area < 30:
        return "25-30"
    elif 30 <= area < 40:
        return "30-35"
    elif 40 <= area < 45:
        return "40-45"
    elif 45 <= area < 50:
        return "35-45"
    elif 50 <= area < 70:
        return "50-70"
    elif 70 <= area < 100:
        return "70-100"
    elif 100 <= area < 175:
        return "100-175"
    elif 175 <= area < 350:
        return "175-350"
    elif 350 < area:
        return "Ну прям огромный"

data['categorized_total_area'] = data['total_area'].apply(categorize_total_area)
```

## Чистим парки

```python
def normal_parks_around3000(parks_around3000: float, city_center: float) -> float:
    if city_center < 3000:
        if parks_around3000 < 1.0: #моя личная додумка - в центре точно есть какой-нибудь парк
            parks_around3000 = 1.0
        if not parks_around3000: #не работает
            parks_around3000 = 1.0
    return parks_around3000

data["parks_around3000"] = data.apply(lambda row: normal_parks_around3000(row["parks_around3000"], row["cityCenters_nearest"]), axis=1)
data["parks_around3000"] = data["parks_around3000"].fillna(0.0)
data
```

```python
data['parks_nearest'] = data.groupby('parks_around3000')['parks_nearest'].transform(lambda x: x.fillna(x.median()))
```
